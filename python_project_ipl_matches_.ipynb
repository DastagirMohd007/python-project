```python
#importing the new library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('/content/matches.csv')
```

```python
#Top five records.
ipl.head()
```

```python
#Check the row and columns in a dataset
ipl.shape
```

```python
#Getting the frequency of man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#Getting top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting top 5 members how received most man of the match
ipl['player_of_match'].value_counts()[0:5]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(8,3))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color= 'g')
plt.xlabel('players')
plt.ylabel('man of the match awards')
plt.title('top 5  man of the match')
plt.show()
```

```python
#making a bar-plot for the top 5 playes with most man of the match awards

plt.figure(figsize=(8,3))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color = 'pink')
plt.xlabel('players')
plt.ylabel('man of the match awards')
plt.title('top 5 players with most man of the match awards')
plt.show()
```

```python
#getting the frequency result column
ipl['result'].value_counts()
```

```python
#finding out the number of toss wins w.r.t each team top 5 team
ipl['toss_winner'].value_counts().head()
```

```python
#extracting the records where a team won batting first
batting_first = ipl[ipl['win_by_runs'] != 0]
batting_first
```

```python
#looking at the head
batting_first.head()
```

```python
#making a histogram
plt.figure(figsize =(8,3))
plt.hist(batting_first['win_by_runs'],bins = 20,color='grey')
plt.xlabel('runs')
plt.ylabel('frequency')
plt.title('distribution of runs')
plt.show()
```

```python
#finding out the number wins w.r.t each team after batting first
batting_first['winner'].value_counts()[0:10]
```

```python
#making a bar-plot for top 3 teams with most wins after batting first

plt.figure(figsize =(8,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color = 'c')
plt.xlabel('teams')
plt.ylabel('wins')
plt.title(' top 3 team with most wins with batting first')
plt.show()
```

```python
#making use of doughnut_Chart chart
plt.figure(figsize=(8,3))
plt.pie(list(batting_first['winner'].value_counts()[0:10]),labels = list(batting_first['winner'].value_counts()[0:10].keys()),autopct='%0.1f%%',radius=4)
plt.pie([1],colors = 'w',radius=2)
plt.show()
```

```python
#extracting those team records where a team has won after second batting

batting_second = ipl[ipl['win_by_wickets'] != 0]
batting_second
```

```python
batting_second.head()
```

```python
#making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize= (4,2))
plt.hist(batting_second['win_by_wickets'],bins =30,color='y')
plt.show()
```

```python
#finding out top 5 the frequency of number of wins w.r.t each time after second batting
batting_second['winner'].value_counts()[0:5]
```

```python
#making a bar plot for top 3 teams with most wins after batting second

plt.figure(figsize=(6,2))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color= 'c')
plt.title('top 3 team won by seconding batting')
plt.xlabel('teams')
plt.ylabel('wins')
plt.show()
```

```python
#amking a pie chart for disturbution of most wins after second batting

plt.figure(figsize=(8,6))
plt.pie(list(batting_second['winner'].value_counts() ), labels= list(batting_second['winner'].value_counts().keys()),autopct= '%0.1f%%',shadow = True)
plt.show()
```
